```python
# Tracer un graphique linéaire
plt.plot([0, 3, 6, 9], [0, 500, 1000, 1500])
plt.xlabel('Quantity ordered')
plt.ylabel('margin')
plt.title('Mon Graphique Linéaire')
plt.show()
```

```python
top_products = data.nlargest(10, 'margin')
```

```python
plt.figure(figsize=(20, 6))  # Adjust the figure size as needed
plt.bar(top_products['Product'], top_products['margin'], color='skyblue', edgecolor='black')
plt.title('Top 2 Products by Margin')
plt.xlabel('Product')
plt.ylabel('Margin')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
import seaborn as sns
```

```python
# Données exemple

# Création d'un diagramme en boîte avec Seaborn
sns.boxplot(x='catégorie', y='margin', data=data)

# Ajout de titres
plt.title('Diagramme en Boîte avec Seaborn')
```

```python
unique_products = data['Product'].unique()

# Afficher les valeurs uniques des produirs
print(unique_products)
```

```python

# Grouping the data by 'Product' and summing the 'Quantity Ordered' for each product
grouped_data = data.groupby('Product')['Quantity Ordered'].sum().reset_index()

# Creating a count plot with Seaborn for the total 'Quantity Ordered' per product in the 'vètements' category
sns.barplot(x='Quantity Ordered', y='Product', data=grouped_data, palette='viridis')

# Adding titles and labels
plt.title('Total Quantity Ordered for Each Product')
plt.xlabel('Total Quantity Ordered')
plt.ylabel('Product')

# Displaying the plot
plt.show()
```
